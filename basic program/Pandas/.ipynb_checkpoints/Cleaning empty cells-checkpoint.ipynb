## iii)Replace using Mean,Median or Mode

```python
data1=pd.read_csv("IMDB-Movie-Data.csv")
x=data1["Revenue (Millions)"].mean()
data1["Revenue (Millions)"].fillna(x,inplace=True)
```

```python
data1.info()
```

```python
y=data1["Metascore"].median()
data1["Metascore"].fillna(y,inplace=True)
```

```python
data1.info()
```

```python
y=data1["Metascore"].mode()
data1["Metascore"].fillna(y,inplace=True)
```
