# 1. You need to produce random numbers from seed using:

```python
count=0
seed=2345

result=[]
while count<1000:
    rand=16807*seed % (2**32 - 1)
    seed=rand
    result.append(rand)
    count+=1
```

```python
import numpy as np
```

```python
arr=np.array(result)
```

```python
a=[(i/np.max(arr)) for i in arr]
```

# MC=monte Carlo

```python
def f(x):
    return np.exp(x)
```

```python
a=2
b=3
N=1000
arr=np.zeros(N)
```

```python
for i in range(len(arr)):
    arr[i]=np.random.uniform(a,b)
```

```python
integral=0.0

for i in arr:
    integral +=f(i)
    
```

```python
mean=integral/N
```

```python
result=(b-a)*mean
```

```python
result
```

## sin

```python
def f2(x):
    return np.sin(x)
```

```python
c=0
d=np.pi
arr2=np.zeros(1000)
```

```python
for i in range(len(arr2)):
    arr2[i]=np.random.uniform(c,d)
```

```python
integral2=0.0

for i in arr2:
    integral2+=f2(i)
```

```python
mean2=integral2/N
```

```python
result2=(d-c)*mean2
```
