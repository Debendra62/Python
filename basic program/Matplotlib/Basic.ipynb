```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4])
b=np.array([2,3,4,1])
```

# Line plot

```python
plt.plot(a,b)
```

# Dotted plot

```python
plt.plot(a,b,"o--r")
```

```python
plt.plot(a,b,"^--r",ms=20)
```

# Stright line plot

```python
c=np.array([1,2,3,4])
d=np.array([1,2,3,4])
```

```python
plt.plot(c,d)
```

# Sub Plot (many plot in one figure)

### Title,show and subplot

```python
a=np.array([1,2,3,4])
b=np.array([2,3,4,1])
plt.subplot(2,1,1)
plt.plot(a,b)
plt.title("plot 1")
plt.show()

c=np.array([1,2,3,4])
d=np.array([1,2,3,4])
plt.subplot(2,1,1)
plt.plot(c,d)
plt.title("plot 2")

plt.show()
```
