# Bargraph (discreate)[x vs y]

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Bar (vertical)

```python
company=["Google","Amazon","Apple","Netflix"]
employee=[1100,2400,3400,6000]
branch=[1000,2000,1400,1100]
plt.bar(company,employee)
plt.bar(company,branch)
```

### Barh (horizontal)

```python
plt.barh(company,employee)
```

### legend( to deferrentiate color )

```python
plt.bar(company,employee)
plt.bar(company,branch)
plt.legend(["employee","branch"])
```

```python
plt.bar(company,employee,label="employee")
plt.bar(company,branch,label="branch")
plt.legend()
```

### width ( to change width )

```python
plt.bar(company,employee,label="employee",width=0.6)
plt.bar(company,branch,label="branch",width=0.6)
plt.legend()
```

### height ( to change height )

```python
plt.barh(company,employee,label="employee",height=0.4)
plt.barh(company,branch,label="branch",height=0.4)
plt.legend()
```

### subplot 

```python
s=np.array([1,2,3,4,7])
p=np.array([3,2,6,1,2])
plt.subplot(2,1,1)
plt.bar(s,p)
plt.title("Plot 1")
plt.show()

x=np.array([1.4,3.34,3.5,2.5,1.9,8.2])
y=np.array([1.5,5.3,1.3,2.4,1.3,2.4])
plt.subplot(2,1,1)
plt.bar(x,y)
plt.title("Plot 2")
plt.show()
```
